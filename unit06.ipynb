**3. 기온 데이터를 히스토그램으로 표현하기**

1) 인천의 기온을 히스토그램으로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('/content/drive/MyDrive/과외/서원이/모두의 데이터분석/첫째마당/Incheon.csv', 'r', encoding = 'cp949') #서원이 파일에 접근
data = csv.reader(f)
next(data)
result = []

for row in data:
  if row[5] != '':
    result.append(float(row[5]))

plt.hist(result, bins = 100, color = 'skyblue')
plt.show()
```

2) 인천의 8월 기온을 히스토그램으로 표현하기

```python
f = open('/content/drive/MyDrive/과외/서원이/모두의 데이터분석/첫째마당/Incheon.csv', 'r', encoding = 'cp949') #서원이 파일에 접근
data = csv.reader(f)
next(data)
aug = []

for row in data:
  if row[5] != '':
    month = row[6].split('-')[1]
    if month == '08':
      aug.append(float(row[5]))

plt.hist(aug, bins = 100, color = 'orange')
plt.show()
```

3) 인천의 1월과 8월 기온을 히스토그램으로 표현하기

```python
f = open('/content/drive/MyDrive/과외/서원이/모두의 데이터분석/첫째마당/Incheon.csv', 'r', encoding = 'cp949') #서원이 파일에 접근
data = csv.reader(f)
next(data)

aug = [] 
jan = []

for row in data:
  if row[5] != '':
    month = row[6].split('-')[1]
    if month == '08':
      aug.append(float(row[5]))
    if month == '01':
      jan.append(float(row[5]))

plt.hist(aug, bins = 100, color = 'red', label = 'Aug')
plt.hist(jan, bins = 100, color = 'blue', label = 'Jan')
plt.legend()
plt.show()
```
