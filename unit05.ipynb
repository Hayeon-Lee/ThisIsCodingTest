**3. 날짜 데이터 추출하기**

```python
f = open('/content/drive/MyDrive/과외/서원이/모두의 데이터분석/첫째마당/Incheon.csv', 'r', encoding = 'cp949') #서원이 파일에 접근
data = csv.reader(f)
next(data)

result2 = []
```

```python
f = open('/content/drive/MyDrive/과외/서원이/모두의 데이터분석/첫째마당/Incheon.csv', 'r', encoding = 'cp949') #서원이 파일에 접근
data = csv.reader(f)
next(data)
result3 = []

for row in data:
  if row[5] !='':
    if row[6].split('-')[1] == '07' and row[6].split('-')[2] == '06': #서원이 생일로 수정해보기
      result3.append(float(row[5]))

plt.plot(result3, 'blue')
plt.show()
```

**4. 내 생일의 기온 변화를 그래프로 그리기**

```python
import csv
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')
```

```python
f = open('/content/drive/MyDrive/과외/서원이/모두의 데이터분석/첫째마당/Incheon.csv', 'r', encoding = 'cp949') #서원이 파일에 접근
data = csv.reader(f)
next(data)

high = [] #최고 기온 값을 저장할 리스트 high 생성
low = [] #최저 기온 값을 저장할 리스트 low 생성

for row in data:
  if row[5] != '' and row[8] != '':#최고기온과 최저기온 값이 존재한다면
    date = row[-1].split('-') #기온 -로 스플릿하기 
    if 1983<=int(date[0]): #1983년 이후 데이터만 
      if date[1] == '07' and date[2] == '06': #서원이 생일로 수정
        high.append(float(row[5])) #최고기온 저장
        low.append(float(row[8])) #최저기온 저장

plt.rcParams['axes.unicode_minus'] = False #마이너스 기호 깨짐 방지
plt.title('Birthday Graph') #제목 설정

plt.plot(high, 'hotpink', label = 'high')
plt.plot(low, 'skyblue', label = 'low')

plt.legend()
plt.show()
```
